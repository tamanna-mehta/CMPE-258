# part b3) Neural network with pytorch (high level primitives)
We are going to make neural network with 3 inputs and 2 outputs. Relu function is used for non linearity at each layer. 2 hidden layers are used -one consisting of 5 neurons and second hidden layer consisting of four neurons.We are tryint to predict more than one value.
we use two non linear functions-
1. 5x1^2+x2^2+x3^2+5x1+x2+x3
2. 4x1^2+x2^2+x3^2+4x1+x2+x3

```
import numpy as np
import random
from numpy import array
from numpy.random import uniform
from numpy import hstack
import torch
import torch.nn as nn
```

```
#creating Non linear model consisting of 3 layers-2 hidden and one output layer consisting of two output neurons.
class TorchModel(nn.Module):
  def __init__(self,input_dim:int,num_hidden1:int,num_hidden2:int):
    super().__init__()
    self.linear1=nn.Linear(input_dim, num_hidden1)
    self.relu1=nn.ReLU()
    self.linear2=nn.Linear(num_hidden1,num_hidden2)
    self.relu2=nn.ReLU()
    self.linear3=nn.Linear(num_hidden2,2)
  
  def __call__(self,x): # forward propagation step
    l1=self.linear1(x)
    r1=self.relu1(l1)
    l2=self.linear2(r1)
    r2=self.relu2(l2)
    l3=self.linear3(r2)
    return l3

  
```

```
n=400  # number of input samples
d=3    # number of dimensions
num_hidden1=5 # number of neurons for first hidden layer
num_hidden2=4 # numbe rof neurons for second hidden layer
x=np.random.uniform(-1,1,(n,d)) # generating data of size=400 X 3

w1_true=np.array([[5,1,1],]).T
w2_true=np.array([[4,1,1],]).T
bias_true=np.array([1])

y1=(x**2) @ w1_true+x @ w1_true+bias_true #final layer output of first neuron 
y2=(x**2) @ w2_true+x @ w2_true+bias_true #final layeroutput of seconf layer neuron
y_true=hstack((y1, y2)) # joining output of first and second neuron
print(f'x: {x.shape},weights:{w1_true.shape}, bias: {bias_true.shape},y:{y_true.shape}')
loss=nn.MSELoss() 
model=TorchModel(d,num_hidden1,num_hidden2) 
x_tensor=torch.tensor(x).float() #converting numpy to tensor
y_true_tensor=torch.tensor(y_true).float() #converting numpy to tensor


torch_fit(x_tensor,y_true_tensor, model=model, loss=loss, lr=0.05, num_epochs=40) # running the non linear model

```

```
def plot_intereactive_3d(x, y, y_pred=None):
  import plotly.graph_objects as go

  fig = go.Figure()
  fig.add_trace(go.Scatter3d(x = x[:,0],
                    y = x[:,1],
                    z = y.reshape([-1]),
                    opacity=0.5, mode='markers', name='Underlying Function'
                    ))
 
  if y_pred is not None:
    fig.add_trace(go.Scatter3d(x = x[:,0],
                   y = x[:,1],
                   z = y_pred.reshape([-1]),
                   opacity=0.5, mode='markers', name='Predicted Function'
                  ))
    
  fig.update_layout(scene = dict(
                    xaxis_title='X1',
                    yaxis_title='X2',
                    zaxis_title='Y'),
                    width=700,
                    margin=dict(r=20, b=10, l=10, t=10))
  fig.show()
```
